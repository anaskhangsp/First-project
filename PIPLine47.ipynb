```python
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps = [("standard_scaler",StandardScaler()),
         ("classifier",LogisticRegression())]
```

```python
pipeline = Pipeline(steps)
pipeline
```

```python
from sklearn import  set_config
set_config(display="diagram")
```

```python
from sklearn.datasets import make_classification
x,y = make_classification(n_samples=1000)
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
pipeline.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = pipeline.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
mse
y_pred
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
umaric_processer = Pipeline(steps=[("imputer",SimpleImputer(missing_values=np.nan,strategy="mean")),
                                     ("scaler",StandardScaler())])
```

```python
from sklearn.preprocessing import OneHotEncoder
Categorical_processer = Pipeline(steps=[("imputer",SimpleImputer(fill_value='missing',strategy="constat")),
                                     ("onehot",OneHotEncoder(handle_unknown="ignor"))])
```

```python
from sklearn.compose import ColumnTransformer
preprocesser = ColumnTransformer(
    [("CAte",Categorical_processer,["geder",'cit']),
     ("num",umaric_processer,["age","eight"])]
)
```
